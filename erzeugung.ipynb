# Stromerzeugung

```python
import pandas as pd
```

```python
%%time
df = pd.read_csv("data/realisierte_erzeugung.csv", na_values='-')
```

```python
df.describe()
```

```python
df.fillna(method='ffill', inplace=True) #replacing '-'(nan-values) by PV and sonstige Erneuerbare with previous value
```

```python
# convert PV and sonstige Erneuerbare to int
df.Photovoltaik = df.Photovoltaik.astype(int)
df['Sonstige Erneuerbare'] = df['Sonstige Erneuerbare'].astype(int)
```

```python
# convert date to pandas datetime
df.Datum = df.Datum + df.Uhrzeit
df['Datum'] = pd.to_datetime(df['Datum'], format="%m/%d/%Y%H:%M")
df.drop(columns='Uhrzeit', inplace = True)
```
